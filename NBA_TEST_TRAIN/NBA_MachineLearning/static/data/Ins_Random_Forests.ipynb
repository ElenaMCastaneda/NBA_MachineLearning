```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
```

```python
shots = pd.read_csv('DBexport_knicks.csv')
shots
```

```python
# Create a random forest classifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(shots.shot_distance, shots.shot_made_flag)
rf.score(shots.shot_distance, shots.shot_made_flag)
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
importances
```

```python
# We can sort the features by their importance
sorted(zip(rf.feature_importances_, shots.feature_names), reverse=True)
```
